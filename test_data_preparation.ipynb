```python
import pandas as pd 
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt 
pd.set_option('display.float_format', lambda x: '%0.4f' % x)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import warnings
warnings.filterwarnings("ignore"
)
```

```python
df =pd.read_csv(r'Datasets\cleaned_and_encoded_test_data.csv')
df.head()
```

```python
df.isnull().sum(axis = 0)
```

```python
# dropping unnamed columns

df=df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)
df.columns
```

```python
# Removing Customers who have not taken loans 

bool_series = pd.notnull(df["No_of_prevloans"])
filtered_data = df[bool_series]
filtered_data.head()
```

```python
filtered_data.isnull().sum(axis = 0)
```

```python
# Dropping columns that are likely causing bias

filtered_data=filtered_data.drop(['bank_name_clients','employment_status_clients','level_of_education_clients'],axis=1)
filtered_data.columns
```

```python
filtered_data.head()
```

```python
# Dropping irrelevant columns
filtered_data=filtered_data.drop(['last_approveddate','last_creationdate','last_closeddate','last_firstrepaiddate','systemloanid'],axis=1)
```

```python
filtered_data=filtered_data.drop(['approveddate','creationdate','last_firstduedate'],axis=1)
```

```python
filtered_data.info()
```

Dropping the columns with high Variance Inflation Factor as per the training

```python
columns_to_drop = ['Total_amount_borrowed','loanamount','totaldue','termdays','bank_account_type_Other','bank_account_type_Savings','loan_close_time']
```

```python
filtered_data=filtered_data.drop(columns_to_drop,axis=1)
```

```python
filtered_data.to_csv(r'Datasets\final_test_data_temp.csv',index=False)
```
