```python
import pandas as pd 
df_test_prev_loans=pd.read_csv('Datasets/testprevloans.csv')
df_test_prev_loans.head()
```

```python
df_test_prev_loans_sum=df_test_prev_loans.groupby('customerid',as_index=False).agg({'systemloanid':'nunique',
                                                                          'loanamount':'sum',
                                                                          'totaldue':'sum',
                                                                          'termdays':'mean',
                                                                          'referredby':pd.Series.mode,
                                                                          'approveddate':'max',
                                                                          'creationdate':'max',
                                                                          'closeddate':'max',
                                                                          'firstduedate':'max',
                                                                          'firstrepaiddate':'max'})
```

```python
df_test_prev_loans_sum.head()
```

```python
df_test_prev_loans_sum.rename(columns={"systemloanid":"No_of_prevloans","loanamount":"Total_amount_borrowed",
                                         "totaldue":"Total_amount_due","termdays":"Average_termdays","referredby":"most_referredby",
                                         "approveddate":"last_approveddate","creationdate":"last_creationdate","closeddate":"last_closeddate",
                                         "firstduedate":"last_firstduedate","firstrepaiddate":"last_firstrepaiddate"},inplace=True)
df_test_prev_loans_sum.head()
```

```python
df_test_demographics=pd.read_csv('Datasets/testdemographics.csv')
df_test_demographics.head()
```

```python
df_test_demographics.customerid.nunique()
```

```python
df_test_demographics.count()
```

There are about 3 duplicate values.

```python
# dropping duplicates 
df_test_demographics=df_test_demographics.drop_duplicates(subset='customerid', keep='first', inplace=False)
len(df_test_demographics)
```

```python
df_test_perf=pd.read_csv('Datasets/testperf.csv')
df_test_perf.head()
```

```python
# merging of the test datasets
df_test_demo_prevloan = pd.merge(df_test_demographics, df_test_prev_loans_sum, how = 'left', left_on = 'customerid', right_on = 'customerid')
df_test_demo_prevloan.head()
```

```python
df_test_final = pd.merge(df_test_demo_prevloan, df_test_perf, how = 'left', left_on = 'customerid', right_on = 'customerid')
df_test_final.head()
```
